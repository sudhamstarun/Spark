```python
import pandas as pd
import math
```

```python
data = pd.read_csv('BuildingNames_Urban.csv', encoding = 'ISO-8859-1')
data2 = pd.read_csv('BuildingNames_NT.csv', encoding = 'ISO-8859-1')
```

```python
count = 0
for j in data['Name of Building']:
    count += 1
count2 = 0
for j in data2['Name of Building']:
    count2 += 1
```

```python
for j in range(0, count):
    str = data.loc[j]['Name of Building']
    newstr = ''
    if(isinstance(str, float)):
        continue
    for i in str:
        if(i != '\n' and i != '*' and i != '?' and i != '\r' and i != '\"'):
            newstr = newstr + i
        elif(i == '*' or i == '?'):
            data.drop([j], axis = 0)
            break
        else:
            break
    data.set_value(j, 'Name of Building', newstr)

for j in range(0, count2):
    str = data2.loc[j]['Name of Building']
    newstr = ''
    if(isinstance(str, float)):
        continue
    for i in str:
        if(i != '\n' and i != '*' and i != '?' and i != '\r' and i != '"'):
            newstr = newstr + i
        elif(i == '*' or i == '?'):
            data2.drop([j], axis = 0)
            break
        else:
            break
    data2.set_value(j, 'Name of Building', newstr)
```

```python
for j in range(0, count):
    str = data.loc[j]['Name of Building']
    word = ''
    flag = True
    if(isinstance(str, float)):
        continue
    for i in str:
        if(i != ' '):
            word = word + i
        else:
            if(word == 'Hotel' or word == 'Church' or word == 'Mall' or word == 'School'):
                flag = False
            word = ''
    if(word == 'Hotel' or word == 'Church' or word == 'Mall' or word == 'School'):
        flag = False
    if(flag):
        data.set_value(j, 'Property_Type', 0)
    else:
        data.set_value(j, 'Property_Type', 1)
        
for j in range(0, count2):
    str = data2.loc[j]['Name of Building']
    word = ''
    flag = True
    if(isinstance(str, float)):
        continue
    for i in str:
        if(i != ' '):
            word = word + i
        else:
            if(word == 'Hotel' or word == 'Church' or word == 'Mall' or word == 'School'):
                flag = False
            word = ''
    if(word == 'Hotel' or word == 'Church' or word == 'Mall' or word == 'School'):
        flag = False
    if(flag):
        data2.set_value(j, 'Property_Type', 0)
    else:
        data2.set_value(j, 'Property_Type', 1)
```

```python
for i in range(0, count):
    str = data.loc[i]['Name of Building']
    if str == '' or str is None:
        data = data.drop([i], axis = 0)

for i in range(0, count2):
    str = data2.loc[i]['Name of Building']
    if str == '' or str is None:
        data2 = data2.drop([i], axis = 0)
```

```python
for i in data.index:
    str = data.loc[i]['Address']
    if str == '' or str is None or str != str:
        data = data.drop([i], axis = 0)

for i in data2.index:
    str = data2.loc[i]['Address']
    if str == '' or str is None or str != str:
        data2 = data2.drop([i], axis = 0)
```

```python
for i in data.index:
    str = data.loc[i]['Address']
    word = ''
    for j in str:
        if j != ' ':
            word = word + j
        else:
            if word == 'Appendix':
                data = data.drop([i], axis = 0)
            word = ''
    if word == 'Appendix':
        data = data.drop([i], axis = 0)

for i in data2.index:
    str = data2.loc[i]['Address']
    word = ''
    for j in str:
        if j != ' ':
            word = word + j
        else:
            if word == 'Appendix':
                data2 = data2.drop([i], axis = 0)
            word = ''
    if word == 'Appendix':
        data2 = data2.drop([i], axis = 0)
```

```python
for i in data.index:
    str = data.loc[i]['Address']
    newstr = ''
    for j in str:
        if j != '\r' and j != '\n':
            newstr = newstr + j
    data.set_value(i, 'Address', newstr)

for i in data2.index:
    str = data2.loc[i]['Address']
    newstr = ''
    for j in str:
        if j != '\r' and j != '\n':
            newstr = newstr + j
    data2.set_value(i, 'Address', newstr)
```
